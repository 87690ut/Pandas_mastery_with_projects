```python
import pandas as pd
```

```python
from dateutil import parser
```

```python
data = pd.read_csv('anime.csv')
data
```

```python
data.drop(columns = ['Count'], inplace= True)
```

```python
data['Episod'] = data['Title'].str.extract(r'\((.*?)\)')  # data extraction called feature extraction.
```

```python
data
```

```python
data['Episod'] = data['Episod'].str.replace('eps','')
```

```python
data['Episod'] = data['Episod'].astype(int)  # changing the value of column type to int
```

```python
pd.set_option('display.max_colwidth',None)
data.iloc[:1,1:2]
```

### Creating a regex pattern for extracting the date anywhere in the Title

```python
pattern = r'([a-zA-Z]{3,}\s\d{4})'
```

### Implement that regex variable to our dataframe column where date is avilable

```python
data['Date'] = data['Title'].str.findall(pattern)
data['Date'] = data['Date'].str.join('-')
data
```

##### Now firstly split seperate date into seperte column then convertit into date time for better operation

```python
data[['Start date','End date']] = data['Date'].str.split('-',expand=True)
```

```python
data
```

### now convert the string formatedate to date time

```python
data['Start date'] = pd.to_datetime(data['Start date'])
data['End date'] = pd.to_datetime(data['End date'])
```

```python
data
```

## Time stamp

#### now find the total months diffrence bw the years or date(for this i already sepearated  the dates into seperate column)

```python
data['Total Months'] = ((data['End date'].dt.year - data['Start date'].dt.year)*12)+(data['End date'].dt.month - data['Start date'].dt.month)+1
```
