# 3. Missing Data

```python
info ={
    'A':[1,np.nan,3,4,5],
    'B':[1,2,3,4,np.nan],
    'C':[1,np.nan,np.nan,np.nan,5]
}
dd = pd.DataFrame(info)
dd
```

#### finding mising data

```python
dd.isna()  # only finding the nan value basis of true or false
dd.isna().sum()  # best way to cunt missing data in all 
```

##### see that all colum have nan or any have not by true and false

```python
# finding that nan isin all column or any colum doesn't contain null
dd.isna().any()
```

### Removing the null values by two way

```python
# Removing the null values - (works on basis of row)
dd.dropna()  # this will remove whole row if a single row find nan in whole row
# dd.dropna(how = 'all')  # this will remove the row if whole row is nan
```

### Thresold, tells how many cells must have to be valid value if not meet that requirment then delete the row

```python
dd.dropna(thresh = 2) # यह रो को तब डिलीट करता है जब कन्डिशन के हिसाब से कम से कम उतनी वैल्यू अगर वैलिड ना हो 
```

## Merging, Joining, Concatinating

```python
employee = pd.DataFrame({
    'employee_id':[1,2,3,4,5],
    'name':['Ajay','Amit','Arjun','Uttam','Sanjay'],
    'department':['HR','IT','Finance','IT','HR']   
})

salaries = pd.DataFrame({
    'employee_id':[1,2,3,6,7],
    'salary':[55000,67000,3400,80000,45000],
    'bonus': [5000,6000,4500,6000,4000]
})
```

```python
employee
```

```python
salaries
```

```python
pd.merge(employee,salaries, on= 'employee_id', how= 'inner') # this will merge both df on the basis of matching employee_id
```

```python
pd.merge(employee,salaries,on='employee_id',how='outer')  # this will oin whole df so join al all and give nan to the cell which id i
#not avilable on each other
```

```python
pd.merge(employee,salaries,on='employee_id',how='left')
```

```python
pd.merge(employee,salaries,on='employee_id',how='right')
```

##  GroupBy Aggregation

```python
grda = pd.DataFrame({
    'category': ['a', 'b', 'a', 'b', 'a', 'b', 'a', 'b'],
    'store': ['s1', 's1', 's2', 's2', 's1', 's2', 's2', 's1'],
    'sales': [100, 150, 200, 175, 120, 160, 180, 190],
    'Quantity': [10, 15, 20, 18, 12, 16, 18, 19],
    'date': pd.date_range('2022-02-01', periods=8)
})

grda
```

```python
an = grda.groupby('category')  
for a, aa in an:
    print(a)
    print(aa)              
```

```python
ans = grda.groupby(['category','store'])[['Quantity','sales']].sum()
print("total sales by category\n",ans)
```

## Aggregation

```python
grda['sales'].mean()  
"""mean mode median min max count std"""
```

```python
grda['sales'].max()
grda[['sales','Quantity']].max()
```

#### multiuple aggrigation at a time

```python
grda[['sales','Quantity']].agg(['min','max','mean'])
```

# Pivot table

```python
ssda = pd.DataFrame({
    'date': pd.date_range('2023-09-01', periods=20),
    'product': ['A', 'B', 'C', 'D'] * 5,
    'region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'] * 2,
    'sales': [5000, 6000, 7500, 4500, 8000, 5500, 6500, 7000, 5800, 6200] * 2,
    'units': [10, 12, 15, 9, 16, 11, 13, 14, 12, 13] * 2,
    'rep': ['John', 'Alice', 'Bob', 'Carol', 'David', 'Eve', 'Frank', 'Grace', 'Henry', 'Iris'] * 2
})
ssda['Month'] = ssda['date'].dt.month_name()
ssda
```

```python
pd.pivot_table(ssda,values = 'sales',index = 'region',columns = 'product',aggfunc = 'count')
```
